```python
clean_df = df.dropna()
clean_df.tail()
```

```python
# The Highest Mid-Career Salary
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmax()]
```

```python
# The Lowest Starting and Mid-Career Salary
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

```python
# Lowest Risk Majors
spread = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', spread)
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Top 5 degrees with the highest values in the 90th percentile
percentile = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
percentile[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Degrees with the greatest spread in salaries
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```
