**Import Pandas**


```python
import pandas as pd
```

**Read CSV and show data**

```python
mhp_ca_df = pd.read_csv('ca_mhp.csv')

mhp_ca_df.shape
```

```python
mhp_ca_df.info()
```

```python
mhp_ca_df.head(10)
```

**Data Clean Up**

```python
# Clean address and phone data
mhp_ca_df[['Street', 'City', 'State_Zip', 'Phone', 'Delete']] = mhp_ca_df['Park Address'].str.split(', ', expand=True)
mhp_ca_df['Address'] = mhp_ca_df['Street'] + ', ' + mhp_ca_df['City'] + ', ' + mhp_ca_df['State_Zip']

# Split operator
mhp_ca_df['Operater'] = mhp_ca_df['Operated by'].str.split(', ', expand=True)[0]

# Delete unneeded columns
columns_to_delete = ['Park Address', 'Delete', 'Operated by', 'Street', 'City', 'State_Zip']
mhp_ca_df = mhp_ca_df.drop(columns=columns_to_delete)

mhp_ca_df.head(10)
```

**Import cleaned data into CSV**

```python
mhp_ca_df.to_csv('updated_ca_mhp.csv', index=False)
```
